```python
# main function- part 1 
def is_safe(levels):
    diffs = []
    for i in range(len(levels) - 1):
        diff = levels[i+1] - levels[i]
        diffs.append(diff)

    all_increasing = True
    for d in diffs:
        if not (1 <= d <= 3):
            all_increasing = False

    all_decreasing = True
    for d in diffs:
        if not (-3 <= d <= -1):
            all_decreasing = False

    if all_increasing or all_decreasing:
        return True
    else:
        return False


# function for part 2
def is_safe_with_dampener(levels):
    # First check if it's already safe
    if is_safe(levels):
        return True

    # Try removing each level one at a time
    for i in range(len(levels)):
        shortened = levels[:i] + levels[i+1:]
        if is_safe(shortened):
            return True

    return False

# # input file opening and converting to 2d array 
reports = []
with open("input.txt") as f:
    for line in f:
        numbers = line.split()
        levels = []
        for n in numbers:
            levels.append(int(n))
        reports.append(levels)

safe_count = 0
for report in reports:
    if is_safe_with_dampener(report):
        safe_count += 1

print(safe_count)
```
